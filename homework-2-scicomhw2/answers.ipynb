1. What is the derivative of `5x^2 + 3x + 1`?

2. Derive a rule for the derivative of $f = g/h$ using symbolic functions.  Does this reduce to quotient rule?

```python
g = functions.Symbolic("g")
h = functions.Symbolic("h")
reciprocal = functions.Power(-1)

one_over_h = functions.Compose(reciprocal, h)
g_over_h = functions.Product(g, one_over_h)

g_over_h_prime = g_over_h.derivative()
print(g_over_h_prime("x"))
```

As shown above, the derivative of $f(x) = \frac{g(x)}{h(x)}$ is $f'(x) = \frac{g'(x)}{h(x)} + \frac{-g(x)h'(x)}{h(x)^2}$. Multiplying the first term by $\frac{h(x)}{h(x)}$ and simplifying, we find that the derivative is $\frac{g'(x)h(x) - g(x)h'(x)}{h(x)^2}$  --- just as the quotient rule states.

3. What is the derivative of $\sin(x)^2$?

```python
sinx = functions.Sin()
sinx_squared = functions.Product(sinx, sinx)
sinx_squared_prime = sinx_squared.derivative()
print(sinx_squared_prime("x"))
```

As shown above, the derivative of $\sin^2(x)$ is $\cos(x)\sin(x) + \sin(x)\cos(x)$, which simplifies to $2\sin(x)\cos(x)$, as expected.

4. What is the second derivative of $e^{5x}$?

```python
expx = functions.Exponential()
fivex = functions.Scale(5)

e_to_fivex = functions.Compose(expx, fivex)

e_to_fivex_prime = e_to_fivex.derivative()
e_to_fivex_double_prime = e_to_fivex_prime.derivative()
print(e_to_fivex_double_prime("x"))
```

As shown above, the second derivative of $e^{5x}$ is $e^{5x} \times 5 \times 5 + e^{5x} \times 0$. Simplifying, we find that the second derivative of $e^{5x}$ is $25e^{5x}$, as expected.
